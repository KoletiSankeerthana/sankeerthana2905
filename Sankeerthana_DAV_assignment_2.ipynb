1) Generate a list of positive integers within a specific range of values (obtained from the user). Find the numbers
from this list which are divisible by 7 and 8 and store them in a separate list using List Comprehension.
Print the new list which contains the numbers divisible by 7 and 8

```python
#Enters the starting value from the user
start=int(input("Enter the starting value:"))
# Enters the ending value from the user
end=int(input("Enter the ending value:"))
#checks weather entered number is in the given range or not
num=[n for n in range(start,end+1)]
# checks weather the range of numbers divisible by 7 and 8 or not
divisible_by_7_and_8=[n for n in num if n%7==0 and n%8==0]
if divisible_by_7_and_8:
    print("List of numbers which are divisible by 7 and 8 in a given range are:",divisible_by_7_and_8)
else:
  print("No numbers which are divisible by both 7 and 8 in a given range")
```

2) Read a string and check if the first character is occurring more than once in the string. If yes, replace the
occurrence by ‘$’ from the second instance. Eg: input: restart, output : resta$t

```python
# Reads  input string from the user
input_str = input("Enter a string: ")

# Takes the first character of the string
frst_char = input_str[0]

# Counts the occurrences of the first character in the string
count = input_str.count(frst_char)

# If the first character occurs more than once, replace by subsequent occurrences
if count > 1:
    replaced_str = frst_char + input_str[1:].replace(frst_char, '$')
else:
    replaced_str = input_str

# Print the replaced string string
print("Replaced string:", replaced_str)
```
